```python
df3 = df2[df2.year != 1970]
```

```python
df3.groupby('year').mean()['price']
```

```python
y = df3['price'].values
x = df3.drop('price',axis=1).values
```

```python
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain,xTest,yTrain,yTest = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
xTrain= scaler.fit_transform(xTrain)
xTest= scaler.transform(xTest)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
xTrain.shape
```

```python
model = Sequential()
model.add(Dense(12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(12, activation='relu'))
# Cikis katmanina activation koymaya gerek yoktur
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=xTrain,y=yTrain,validation_data=(xTest,yTest),batch_size=250,epochs=300)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head(10)
```

```python
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```
