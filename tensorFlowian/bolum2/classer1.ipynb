```python
y = df['Type']
x = df.drop('Type',axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain,xTest,yTrain,yTest = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(xTrain)
```

```python
xTrain = scaler.transform(xTrain)
```

```python
xTest = scaler.transform(xTest)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
xTrain.shape
```

```python
model = Sequential()
# first dens should be equal to number of features
model.add(Dense(units =30, activation='relu'))
# hidden layers should be between input and output
model.add(Dense(units =15, activation='relu'))
model.add(Dense(units =15, activation='relu'))
# last layer should be equal to number of classes
model.add(Dense(units =1, activation='sigmoid'))
# compile the model
model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
# epochs should be enough to train the model but not too much or the model overfits
model.fit(x=xTrain,y=yTrain,epochs=700,validation_data=(xTest,yTest),verbose=1)
```

```python
mK = pd.DataFrame(model.history.history)
mK.plot()
```

```python
model1 = Sequential()
# first dens should be equal to number of features
model1.add(Dense(units =30, activation='relu'))
# hidden layers should be between input and output
model1.add(Dense(units =15, activation='relu'))
model1.add(Dense(units =15, activation='relu'))
# last layer should be equal to number of classes
model1.add(Dense(units =1, activation='sigmoid'))
# compile the model
model1.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
EaS = EarlyStopping(monitor='val_loss',mode='min',patience=25)
```

```python
model1.fit(x=xTrain, y=yTrain, epochs=700, validation_data=(xTest,yTest),verbose=1,callbacks=[EaS])
```

```python
mK = pd.DataFrame(model1.history.history)
mK.plot()
```

```python
model3 = Sequential()
# first dens should be equal to number of features
model3.add(Dense(units =30, activation='relu'))
model3.add(Dropout(0.6))
# hidden layers should be between input and output
model3.add(Dense(units =15, activation='relu'))
model3.add(Dropout(0.6))
model3.add(Dense(units =15, activation='relu'))
model3.add(Dropout(0.6))
# last layer should be equal to number of classes
model3.add(Dense(units =1, activation='sigmoid'))
# compile the model
model3.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model3.fit(x=xTrain, y=yTrain, epochs=700, validation_data=(xTest,yTest),verbose=1,callbacks=[EaS])
```

```python
mK = pd.DataFrame(model3.history.history)
mK.plot()
```

```python
tahminler = model3.predict(xTest)
tahminler
```

```python
y_class = (tahminler > 0.5).astype(int)
```

```python
y_class
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(yTest, y_class))
```

```python
print(confusion_matrix(yTest, y_class))
```
